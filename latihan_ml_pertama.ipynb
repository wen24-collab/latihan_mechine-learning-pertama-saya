```python
#memperdiksi jeni bunga berdasarkan fitur bunga 
#mengimport sklearn

#baris ini mnyedidakan data latihan

from sklearn.datasets import load_iris #kita menggunakan data set bugan iris
#meisahkan data latih dan data uji

from sklearn.model_selection import train_test_split 
#disini model mengkalsifikasikan data bunga iris

from sklearn.linear_model import LogisticRegression #liner ini betuk belajar nya logistic regression cara belajarny
#disni kita melihat nilai akurasi model

from sklearn.metrics import accuracy_score
#kita load data bunga iris

data_bunga_iris = load_iris() #var data_bunga_iris menyimpan fitur label dan nama class dari data
fitur = data_bunga_iris.data #ini fitur(data
label = data_bunga_iris.target #ini label(label data)

#split data latih dan data uji
x_latih, x_uji, y_latih, y_uji = train_test_split(fitur, label, test_size=0.3, random_state=42)

#disni kita buat model nya belajar

model = LogisticRegression(max_iter= 200) #max_iter=200 artinya model akan melakukan iterasi sebanyak 200 kali untuk mencari solusi terbaik

#kita latih model
model.fit(x_latih, y_latih)

#perdiksi data uji
y_pred = model.predict(x_uji)

#disini kita lihat akursi dari model

akurasi_model = accuracy_score(y_uji, y_pred)
print(f'akurasi dari model {akurasi_model}')


```
